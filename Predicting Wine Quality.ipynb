```python
#importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# Dataset details

## importing the dataset

```python
#importing dataset
wine = pd.read_csv("winequality.csv")
wine.head()
```

```python
#copying the original data
data = wine.copy()
data.info()
```

```python
#renaming the columns
data.rename(columns={'fixed acidity':'fixed_acidity','volatile acidity':'volatile_acidity','citric acid':'citric_acid',
                     'residual sugar':'residual_sugar','free sulfur dioxide':'free_sulfur_dioxide',
                     'total sulfur dioxide':'total_sulfur_dioxide'},inplace=True)
```

```python
#checking missing values
data.isna().sum()
```

```python
data.describe()
```

## Imputation of Missing values

```python
#null values are filled with median 
data.fixed_acidity.fillna(data.fixed_acidity.median(),inplace=True)
data.volatile_acidity.fillna(data.volatile_acidity.median(),inplace=True)
data.citric_acid.fillna(data.citric_acid.median(),inplace=True)
data.residual_sugar.fillna(data.residual_sugar.median(),inplace=True)                             
data.chlorides.fillna(data.chlorides.median(),inplace=True)
data.pH.fillna(data.pH.median(),inplace=True)
data.sulphates.fillna(data.sulphates.median(),inplace=True)
```

```python
#checking number of unique values
data.nunique()
```

# Exploratory data analysis (EDA)

## EDA of Variable

```python
#EDA of variables
for i in data.columns[1:-1]:
    print('Variable_Name : ',i.upper())
    fig, axes = plt.subplots(1,2,figsize=(15,5))
    axes[0].set_title('Distribution')
    data[i].hist(ax=axes[0])
    axes[1].set_title('Checking outliers')
    sns.boxplot(data[i],ax=axes[1])
    plt.show()
```

```python
#visualizing categorical variables
sns.countplot(data.type)
```

```python
sns.countplot(data.quality)
```

```python
#checking corelation
plt.figure(figsize=(15,6))
sns.heatmap(data.corr(),annot=True,cmap='RdBu')
```

The variables Total sulphur dioxide and free sulfur dioxide are 72% positively corelated

```python
#visualizing how the type variable is related to input variables
for i in data.columns[1:-1]:
    plt.figure(figsize=(15,6))
    sns.boxplot(x=data.type,y=data[i])
    plt.title('Relationship b/w {} and type'.format(i))
    plt.show()
```

```python
#visualizing how the output is related to input variables
for i in data.columns[1:-1]:
    plt.figure(figsize=(15,6))
    sns.boxplot(x=data.quality,y=data[i])
    plt.title('Relationship b/w {} and quality'.format(i))
    plt.show()
```

```python
pd.crosstab(data.quality,data.type)
```

```python
#one hot encoding the type variable
data.type.replace({'red':0,'white':1},inplace=True)
```

```python
#splitting the data into input and output
x = data.drop('quality',axis=1)
y = data.quality
print('Shape of X = {}, Y = {}'.format(x.shape,y.shape))
```

Since output variable distribution is uneven. Over sampling technique is used

## handling the unbalanced data

```python
#importing the libraries
from collections import Counter
from imblearn.over_sampling import RandomOverSampler

#initializing the over sampling
sampler = RandomOverSampler()

#Over sampled data
x1, y1 = sampler.fit_resample(x,y)

print('Actual y_train count :',Counter(y))
print('y_train After Over Sampling :',Counter(y1))
```

# Feature Selection

```python
#Checking for feature importance using random forest feature importance
#importing libraries
from sklearn.ensemble import RandomForestClassifier

#initializing the model
rand = RandomForestClassifier()

#fitting the model
rand.fit(x1,y1)

#plotting frature importance
plt.title('Random Forest Feature Importance')
sns.barplot(x=rand.feature_importances_,y=x1.columns,color='Blue')
plt.show()
```

```python
#Checking for feature importance using Decision tree feature importance
#importing libraries
from sklearn.tree import DecisionTreeClassifier

#initializing the model
deci = DecisionTreeClassifier()

#fitting the model
deci.fit(x1,y1)

#plotting frature importance
plt.title('CART feature importance')
sns.barplot(x=deci.feature_importances_,y=x1.columns,color='Blue')
plt.show()
```

```python
#from above graph and corelation graph type and free sulfer dioxide variable is removed
#splitting the data into train and test
x1.drop(['type','free_sulfur_dioxide'],axis=1,inplace=True)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1,y1,test_size=0.2,random_state=2)
print('Shape of Splitting:')
print('x_train = {}\nx_test = {}\n\ny_train = {}\ny_test = {}'.format(x_train.shape,x_test.shape,y_train.shape,y_test.shape))
```

# Building the Model

## Building the logistic regression model

```python
#importing the model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, plot_confusion_matrix

#initializing th model
logis = LogisticRegression(solver='liblinear')

#fitting the model and predicting for test data
pred_logis = logis.fit(x_train,y_train).predict(x_test)

#plotting the confusion matrix
print('CONFUSION MATRIX : ')
fig, ax = plt.subplots(figsize=(15, 8))
plot_confusion_matrix(estimator=logis, X=x_test, y_true=y_test,cmap='YlGn_r',ax=ax)
plt.show()

#printing the classification report
print('REPORT: ',classification_report(y_pred=pred_logis,y_true=y_test))

#calculating accuracy
acc_logis = accuracy_score(y_test,pred_logis)
```

```python
#checking hyper-parameters
logis.get_params().keys()
```

```python
#Hyper parameter tuning using GridSearchCV
from sklearn.model_selection import GridSearchCV

#hyper parameters are
params = {'solver' : ['liblinear'],
          'penalty' : ['l2','l1'],
          'C' : [100, 10, 1.0]
         }

#initializing the grid
grid = GridSearchCV(estimator=logis,param_grid=params,cv=3,verbose=3,n_jobs=-1)

#fitting the model
grid.fit(x_train,y_train)

print('Best Score : ',grid.best_score_)
print('Best parameters : ',grid.best_params_,'\n')

#predicting for test data
pred_logis = grid.predict(x_test)

#plotting the confusion matrix
print('CONFUSION MATRIX : ')
fig, ax = plt.subplots(figsize=(15, 8))
plot_confusion_matrix(estimator=grid, X=x_test, y_true=y_test,cmap='YlGn_r',ax=ax)
plt.show()

#printing the classification report
print('REPORT: ',classification_report(y_pred=pred_logis,y_true=y_test))

#calculating accuracy
acc_logis = accuracy_score(y_test,pred_logis)
```

## Building the KNN model

```python
#importing libraries
from sklearn.neighbors import KNeighborsClassifier

#initializing th model
knn = KNeighborsClassifier()

#fitting the model and predicting for test data
pred_knn = knn.fit(x_train,y_train).predict(x_test)

#plotting the confusion matrix
print('CONFUSION MATRIX : ')
fig, ax = plt.subplots(figsize=(15, 8))
plot_confusion_matrix(estimator=knn, X=x_test, y_true=y_test,cmap='YlGn_r',ax=ax)
plt.show()

#printing the classification report
print('REPORT: ',classification_report(y_pred=pred_knn,y_true=y_test))

#calculating accuracy
acc_knn = accuracy_score(y_test,pred_knn)
```

```python
#hyper-parameters
knn.get_params().keys()
```

```python
#hyper parameters are
params = {'n_neighbors' : range(2,15),
          'leaf_size' : range(1,15)
         }

#initializing the grid
grid = GridSearchCV(estimator=knn,param_grid=params,cv=3,verbose=3,n_jobs=-1)

#fitting the model
grid.fit(x_train,y_train)

print('Best Score : ',grid.best_score_)
print('Best parameters : ',grid.best_params_,'\n')

#predicting for test data
pred_knn = grid.predict(x_test)

#plotting the confusion matrix
print('CONFUSION MATRIX : ')
fig, ax = plt.subplots(figsize=(15, 8))
plot_confusion_matrix(estimator=grid, X=x_test, y_true=y_test,cmap='YlGn_r',ax=ax)
plt.show()

#printing the classification report
print('REPORT: ',classification_report(y_pred=pred_knn,y_true=y_test))

#calculating accuracy
acc_knn = accuracy_score(y_test,pred_knn)
```

## Building SVM model

```python
#importing the libraries
from sklearn.svm import SVC

#initializing th model
svm = SVC()

#fitting the model and predicting for test data
pred_svm = svm.fit(x_train,y_train).predict(x_test)

#plotting the confusion matrix
print('CONFUSION MATRIX : ')
fig, ax = plt.subplots(figsize=(15, 8))
plot_confusion_matrix(estimator=svm, X=x_test, y_true=y_test,cmap='YlGn_r',ax=ax)
plt.show()

#printing the classification report
print('REPORT: ',classification_report(y_pred=pred_svm,y_true=y_test))

#calculating accuracy
acc_svm = accuracy_score(y_test,pred_svm)
```

## Building the Adaboost model

```python
#importing libraries
from sklearn.ensemble import AdaBoostClassifier

#initializing th model
ada = AdaBoostClassifier()

#fitting the model and predicting for test data
pred_ada = ada.fit(x_train,y_train).predict(x_test)

#plotting the confusion matrix
print('CONFUSION MATRIX : ')
fig, ax = plt.subplots(figsize=(15, 8))
plot_confusion_matrix(estimator=ada, X=x_test, y_true=y_test,cmap='YlGn_r',ax=ax)
plt.show()

#printing the classification report
print('REPORT: ',classification_report(y_pred=pred_ada,y_true=y_test))

#calculating accuracy
acc_ada = accuracy_score(y_test,pred_ada)
```

```python
#Hyper-parameters
ada.get_params().keys()
```

```python
#hyper parameters are
params = {'n_estimators' : [100,200,300],
          'learning_rate':[0.001,0.01,0.1]
         }

#initializing the grid
grid = GridSearchCV(estimator=ada,param_grid=params,cv=3,verbose=3,n_jobs=-1)

#fitting the model
grid.fit(x_train,y_train)

print('Best Score : ',grid.best_score_)
print('Best parameters : ',grid.best_params_,'\n')

#predicting for test data
pred_ada = grid.predict(x_test)

#plotting the confusion matrix
print('CONFUSION MATRIX : ')
fig, ax = plt.subplots(figsize=(15, 8))
plot_confusion_matrix(estimator=grid, X=x_test, y_true=y_test,cmap='YlGn_r',ax=ax)
plt.show()

#printing the classification report
print('REPORT: ',classification_report(y_pred=pred_ada,y_true=y_test))

#calculating accuracy
acc_ada = accuracy_score(y_test,pred_ada)
```
